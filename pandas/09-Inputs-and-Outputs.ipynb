## SQL

sqlalchemy : https://docs.sqlalchemy.org/en/20/

```python
import numpy as np
import pandas as pd
```

```python
from sqlalchemy import create_engine
```

```python
temp_db = create_engine('sqlite:///:memory:')
```

```python
df = pd.DataFrame(data=np.random.randint(low=0, high=100, size =(4,4)),columns=['a','b','c','d'])
```

```python
df
```

```python
df.to_sql(name='new_table',con=temp_db, index=False)
```

```python
new_df = pd.read_sql(sql='new_table',con=temp_db)
```

```python
new_df
```

```python
result = pd.read_sql_query(sql='SELECT * FROM new_table',con=temp_db)
result
```

### Connect to mylocal database

```python
engine = create_engine("mysql+pymysql://root:@localhost:3306/cyclistic?charset=utf8mb4")
```

```python
df = pd.read_sql_table('202101-divvy-tripdata',engine)
```

```python
df
```

```python
df = pd.read_sql_table('202101-divvy-tripdata',engine, columns=['rideable_type','ride_id'])
df
```

```python
query = '''
    SELECT * 
    FROM `202101-divvy-tripdata` 
    WHERE rideable_type = 'classic_bike'
'''

result = pd.read_sql_query(query,engine)
result
```

```python
query = '''
    SELECT 
        rideable_type,
        COUNT(rideable_type) AS count_rideable_type
    FROM `202101-divvy-tripdata` 
    GROUP BY rideable_type
'''

result = pd.read_sql_query(query,engine)
result
```
