```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Sales_Funnel_CRM.csv')
```

```python
df
```

```python
licenses = df[['Company','Product','Licenses']]
licenses
```

```python
pd.pivot(data=licenses,index='Company',columns='Product', values = 'Licenses')
```

```python
df
```

```python
pd.pivot_table(df,index='Company',aggfunc='sum', values=['Licenses','Sale Price'])
```

```python
pd.pivot_table(df,index=['Account Manager', 'Contact'],columns=['Product'],aggfunc='sum', values=['Sale Price'], fill_value=0)
```

```python
pd.pivot_table(df,index=['Account Manager', 'Contact'],
                columns=['Product'],
                aggfunc=[np.sum, np.mean], 
                values=['Sale Price'], 
                fill_value=0,
                margins=True,
                margins_name='Total')
```

```python
df.groupby('Company').sum()[['Licenses','Sale Price']].sort_values(by = ['Licenses','Sale Price'],ascending=[False,False])
```
