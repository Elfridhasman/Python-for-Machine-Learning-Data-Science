```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
# plt.figure(figsize=(2,2),dpi=200)
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
# sns.set(style='whitegrid')
sns.displot(data=df,x='salary',color='#0cb0c9',edgecolor='white',linewidth=1,kde=True,rug=True)
```

```python
sns.histplot(data=df,x='salary',bins=10,color='#0cb0c9',edgecolor='white',linewidth=1)
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
sns.kdeplot(data=sample_ages,x='age',clip=[0,100],bw_adjust=0.5,shade=True)
```
