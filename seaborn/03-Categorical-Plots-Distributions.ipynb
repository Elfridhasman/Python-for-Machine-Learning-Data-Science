```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df, x='level of education',hue='division',palette='Set2')
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(9,4),dpi=200)
sns.barplot(data=df, x='level of education',y='salary',estimator=np.mean,ci='sd', hue='division')
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
round(df.groupby('level of education').agg({'salary':['mean','std']}),2)
```

```python
student_performance = pd.read_csv('StudentsPerformance.csv')
```

```python
student_performance.head()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=student_performance,y='math score',x='test preparation course')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=student_performance,y='reading score',x='parental level of education', 
            hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.boxplot(data=student_performance, x='reading score',y='parental level of education',
            hue='test preparation course',palette='Set2')
            
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=student_performance, x='reading score',y='parental level of education',
            hue='test preparation course',palette='Set2',split=True,inner=None)
            
plt.legend(bbox_to_anchor=(1.20,1))
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=student_performance, x='reading score',y='parental level of education',
            hue='test preparation course',palette='Set2',split=True,inner='quartile')
            
plt.legend(bbox_to_anchor=(1.20,1))
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=student_performance, x='reading score',y='parental level of education',
               split=True,inner='quartile')
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=student_performance, x='reading score',y='parental level of education',
               split=True,inner='stick')
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=student_performance, x='reading score',y='parental level of education',
               bw=0.1)
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(data=student_performance,x='math score',y='gender',
            hue='test preparation course',dodge=True,size=2)
plt.legend(bbox_to_anchor=(-0.01,1))
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.boxenplot(data=student_performance,x='math score',y='test preparation course',
            hue='gender')
```
