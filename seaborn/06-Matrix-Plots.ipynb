```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('country_table.csv')
```

```python
df = df.set_index('Countries')
```

```python
df
```

```python
plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy',axis=1), linewidths=0.8,annot=True)
```

```python
plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy',axis=1), linewidths=0.8,annot=True,
                   cmap='viridis',center=40)
```

```python
plt.figure(dpi=200)
sns.clustermap(df.drop('Life expectancy',axis=1), linewidths=0.8,annot=True,
                   cmap='viridis',col_cluster=False)
```
