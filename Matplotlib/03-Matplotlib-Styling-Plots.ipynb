```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, label='X vs X')
ax.plot(x,x**2, label='X vs X^2')
# ax.legend(loc='upper left')
ax.legend(loc=(-0.3,0.5))
# ax.legend(loc=0)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.plot(x,x,color='#137378', label='x vs x',lw=5)
# ax.plot(x,x+1,color='purple', label='x vs x+1',lw=3,ls='--')
ax.legend()

lines  = ax.plot(x,x,color='#137378',lw=5)
lines[0].set_dashes([1,1,1,2,3,5])
```
